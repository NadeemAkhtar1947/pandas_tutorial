# **Pandas Tutorial**

#### **What is Python Pandas ?**


*  Pandas is a powerful and open-source Python library that simplifies the tasks of data manipulation in Python.
*   Pandas is well-suited for working with tabular data, such as spreadsheets or SQL tables.

* Pandas name is derived from “panel data” and is also refered as “Python Data Analysis“.

##### **Data structures in pandas**

* Pandas provides two types of classes for handling data:
* Series: a 1-D labeled array holding data of any type
such as integers, strings, Python objects etc.

* DataFrame: a 2-D data structure that holds data like a two-dimension array or a table with rows and columns.

##### 1. **Series**

* Series is a 1-D labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.).
* The basic method to create a Series is to call:
* s = pd.Series(data, index=index)


* Here, data can be many different things:

1. a Python dict
2. an ndarray
3. a scalar value (like 5)

#### **Creating a Series**

```python
### 1. From a List
import pandas as pd
data = [10, 20, 30, 40, 50]
series_from_list = pd.Series(data)
print(series_from_list)
```

```python
### 2. From a NumPy Array

import pandas as pd
import numpy as np
data = np.array([10, 20, 30, 40, 50])
series_from_array = pd.Series(data)
print(series_from_array)
```

```python
### 3. From a Dictionary
import pandas as pd
data = {'a': 10, 'b': 20, 'c': 30, 'd': 40, 'e': 50}
series_from_dict = pd.Series(data)
print(series_from_dict)
```

```python
### 4. From Scalar Value
import pandas as pd
value = 5
length = 5
index_labels = ['a', 'b', 'c', 'd', 'e']
series_from_scalar = pd.Series(value, index=index_labels)
print(series_from_scalar)
```

```python
### 5. From a Tuple
import pandas as pd
data = (10, 20, 30, 40, 50)
series_from_tuple = pd.Series(data)
print(series_from_tuple)
```

```python
### 6. From a range
import pandas as pd
start = 0
stop = 10
step = 2
series_from_range = pd.Series(range(start, stop, step))
print(series_from_range)
```

### **summarize data in series**

```python
import pandas as pd
import numpy as np

data = {'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5, 'F': np.nan}
series = pd.Series(data)

# Basic statistics summary
summary = series.describe()

# Count of non-null values
count_non_null = series.count()

# Sum of all values (excluding NaN)
total_sum = series.sum()

# Mean value
mean_value = series.mean()

# Median value
median_value = series.median()

# Standard deviation
std_deviation = series.std()

# Min and Max values
min_value = series.min()
max_value = series.max()

print("Basic statistics summary:")
print(summary)
print("\nCount of non-null values:", count_non_null)
print("Sum of all values (excluding NaN):", total_sum)
print("Mean value:", mean_value)
print("Median value:", median_value)
print("Standard deviation:", std_deviation)
print("Min value:", min_value)
print("Max value:", max_value)
```

**value_counts():** This method returns the frequency of unique values in the Series.

```python
data = {'A': 3, 'B': 2, 'C': 3, 'D': 2, 'E': 5, 'F': np.nan}
series = pd.Series(data)
print(series.value_counts())
```

```python
data = {'A': 10, 'B': 12, 'C': 23, 'D': 12, 'E': 10, 'F': 43}
series = pd.Series(data)
print("unique values:")
print(series.unique())

# nunique_values(): This method returns the no of unique values in the Series.
print("no of unique values:")
print(series.nunique())

# Sorting the values using sort_values()
print("sorted values in series:")
print(series.sort_values())

print("changed datatype into float:")
print(series.astype(float))

# return max valued index
print(series.idxmax())

# return min valued index
print(series.idxmin())

# convert series into list
print(series.tolist())
```

### **Handling missing data in series**


* Identify Missing Data:
Use isnull() or notnull() methods to identify missing values.
* Remove Missing Data:
Use dropna() method to remove rows with missing values.
* Replace Missing Data:
Use fillna() method to replace missing values with a specified value.

```python
data = {'A': 1, 'B': 2, 'C': 3, 'D': np.nan, 'E': 5, 'F': np.nan, 'G':12, 'H':33}
series = pd.Series(data)

# 1. Identify missing data
print(series.isnull())  # return False when not null and True when null

print(series.notnull())  # return True when not null and False when null

# 2. Remove missing data
print(series.dropna())

# 3. Replace missing data
print(series.fillna(0))    # Replace missing values with anything(here 0)
```

#### **Grouping Data in Series**

```python
import pandas as pd
data = {'Category': ['A', 'B', 'C', 'B', 'A', 'B'],
        'Values': [1, 2, 3, 4, 5, 6]}
series = pd.Series(data['Values'], index=data['Category'])

# Group by index (Category in this case) and calculate the sum
grouped_series = series.groupby(level=0).sum()

print("Grouped Series:")
print(grouped_series)
```

##### **a). Creating a Pandas DataFrame**


* Pandas DataFrame will be created by loading the datasets from existing
storage, storage can be SQL Database, CSV file, and Excel file.
* Pandas DataFrame can be created from the lists, dictionary, and from a list of dictionary etc.

```python
# 1. Creating a dataframe using single list or a list of lists.
import pandas as pd
data = [
    ['Alice', 25, 'New York'],
    ['Bob', 30, 'Los Angeles'],
    ['Charlie', 35, 'Chicago'],
    ['David', 40, 'Houston'],
    ['Emily', 45, 'Phoenix']
]

# Create a DataFrame from the list of lists
df = pd.DataFrame(data, columns=['Name', 'Age', 'City'])
df
```

**zip():** It is a built-in Python function that takes iterables (such as lists, tuples, or strings) as arguments and returns an iterator of tuples where the i-th tuple contains the i-th element from each of the input iterables.

```python
# 4. create dataframe from lists using zip()
import pandas as pd
list1 = ['India', 'South Africa', 'England', 'New Zealand', 'Australia']
list2 = ['England', 'India', 'New Zealand', 'South Africa', 'Pakistan']
list3 = ['Pakistan', 'India', 'Australia', 'England', 'New Zealand']

# Using zip to combine the three lists
zipped = zip(list1, list2, list3)

# Converting the zip object to a list
data = list(zipped)

# Creating DataFrame from the zipped data
df = pd.DataFrame(data, columns=['Test', 'ODI', 'T20'])
df
```

##### **c). header parameter**
It specify which row of the input file should be considered as the column names (header) of the DataFrame when reading data from external sources like CSV, Excel, or SQL databases.

The header parameter can take several different forms:

* Integer: If header is an integer, it indicates the row(s) (0-indexed) to use as the header. e.g , header=0 specifies that the first row of the input file contains the column names.
* None: If header=None, pandas will automatically generate column names as integers starting from 0.
* 'infer': If header='infer', pandas automatically detect the header row by examining the first few rows of the input file.
* List of Strings: If header is a list of strings, it explicitly specifies the column names to use. This is useful when the input file doesn't have a header row.

```python
import pandas as pd
# Read CSV file into DataFrame
df = pd.read_csv('/content/Top 100 Greatest Hollywood Actors of All Time.csv', header=0)
df.head(2)
```

#### **9. Merging in DataFrame:**
* merge() function is used for combining DataFrames based on one or more key columns and more versatile than join.
* when the column names are the same in both DataFrames then use **on** parameter to specify the common key column.
* When the column names are different, you can use left_on and right_on parameters to specify the columns from the left and right DataFrames, respectively.
* By default, merge() performs a inner join and retains only the rows with matching keys from both DataFrames.
* You can specify different types of joins using the how parameter ('inner', 'outer', 'left', 'right').




* **9.1 Inner Join:** Collects only the common data from both the left and right dataframes. This more like an intersection.
* result = pd.merge(df1, df2, on='key_column', how='inner')  # Inner join based on key column

```python
import pandas as pd
# Create two sample DataFrames
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'],
                    'value1': [1, 2, 3, 4]})
df2 = pd.DataFrame({'key': ['B', 'D', 'E', 'F'],
                    'value2': [5, 6, 7, 8]})
# Inner Join
inner_join = pd.merge(df1, df2, on='key', how='inner')
print("Inner Join:")
inner_join
```

* **9.2 Left Join:** Collects all data from left dataframe and common data of left and right.
* result = pd.merge(df1, df2, on='key_column', how='left')  # Left join

```python
# Left Join
left_join = pd.merge(df1, df2, on='key', how='left')
print("Left Join:")
left_join
```

* **9.3 Right Join:** Collects all data from right dataframe and common data of left and right.
* result = pd.merge(df1, df2, on='key_column', how='right')  # Right join

```python
# Right Join
right_join = pd.merge(df1, df2, on='key', how='right')
print("Right Join:")
right_join
```

* **9.4 Outer Join:** Collects all from left and right including common.
* result = pd.merge(df1, df2, on='key_column', how='outer')  # Outer join

```python
# Outer Join
outer_join = pd.merge(df1, df2, on='key', how='outer')
print("Outer Join:")
outer_join
```

#### **10. Joining in DataFrame:**
* join() method is used to join two DataFrames based on **their indexes**.
* By default, it performs a **left join** and retains all rows from the left DataFrame.
* You can specify different types of joins using the how parameter ('inner', 'outer', 'left', 'right').
* It's more limited compared to merge() in terms of customization, especially when joining on columns with different names.


* **10.1 Left Join:**  Collects all data from left dataframe and common data of left and right.
* result = df1.join(df2, how='left')  

```python
import pandas as pd
left_df = pd.DataFrame({'A': [1, 2, 3]}, index=['a', 'b', 'c'])
right_df = pd.DataFrame({'B': [4, 5, 6]}, index=['a', 'b', 'd'])

# Left join
left_join = left_df.join(right_df, how='left')
print("Left Join:")
left_join
```

* **10.2 Right Join:** Collects all data from right dataframe and common data of left and right.
* result = df1.join(df2, how='right')  # right join based on indices

```python
# Right join
right_join = left_df.join(right_df, how='right')
print("Right Join:")
right_join
```

* **10.3 Inner Join:** Collects only the common data from both the left and right dataframes
* result = df1.join(df2, how='inner')  # inner join based on indices

```python
# Inner join
inner_join = left_df.join(right_df, how='inner')
print("Inner Join:")
inner_join
```

* **10.4 Outer Join:** Collects all from left and right including common.
* result = df1.join(df2, how='outer')  # outer join based on indices

```python
# Outer Join
outer_join = left_df.join(right_df, how='outer')
print("Outer Join:")
outer_join
```

A box plot conveys useful information, such as the interquartile range (IQR), the median, and the outliers of each data group.
